## 3. ***Data Wrangling***

### Data Wrangling Code

````python
data=df.copy()
data['Date']=data['Date'].apply(lambda x: datetime.datetime.strptime(x,'%b-%y'))
data.head()
# data.info()

````

````python
data.set_index('Date', inplace=True)
data.sort_index(inplace=True)
````

````python
data['Price_Spread'] = data['High'] - data['Low']
data['Month_Return'] = data['Close'] - data['Open']
````

````python
data['Prev_Month_Close'] = data['Close'].shift(1)
data['Prev_2_Month_Close'] = data['Close'].shift(2)
data['Prev_3_Month_Close'] = data['Close'].shift(3)

data.dropna(inplace=True)
````

````python
data['Log_Open'] = np.log10(data['Open'])
data['Log_High'] = np.log10(data['High'])
data['Log_Low'] = np.log10(data['Low'])
data['Log_Close'] = np.log10(data['Close'])
````

````python
data['Is_Post_Crisis'] = np.where(data.index >= '2018-09-01', 1, 0)
````

````python
data.info()
````

### What all manipulations have you done and insights you found?

1. Chronological Alignment: Indexed and sorted records by timeline to preserve temporal continuity.
2. Risk & Volatility Quantified: Generated intra-month price spreads ($\text{High} - \text{Low}$) to track uncertainty.
3. Sequence Memory Added: Engineered historical lag features (Prev_Month_Close) to capture long-term momentum.
4. Distribution Normalized: Solved heavy right-skewness and stabilized variance using Log transformations.
5. Contextual Regime Flagging: Created a binary operational flag dividing the data into pre- and post-2018 crisis periods.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Hypothetical Statement - 1: The mean monthly closing price of Yes Bank stock dropped significantly after the corporate governance crisis of September 2018 compared to its operational history before the scandal.
- Null Hypothesis ($H_0$): There is no significant difference in the mean monthly closing price before and after the September 2018 crisis ($\mu_{\text{pre}} = \mu_{\text{post}}$). Any observed drop is due to random market fluctuations.
- Alternate Hypothesis ($H_1$): The mean monthly closing price after the September 2018 crisis is significantly lower than the mean monthly closing price before the crisis ($\mu_{\text{pre}} > \mu_{\text{post}}$).

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats
pre_crisis = data[data.index < '2018-09-01']['Close']
post_crisis = data[data.index >= '2018-09-01']['Close']
t_stat, p_value = stats.ttest_ind(pre_crisis, post_crisis, equal_var=False, alternative='greater')
print("--- Hypothesis Testing Report ---")
print(f"Pre-Crisis Sample Size: {len(pre_crisis)} months | Mean Price: INR {pre_crisis.mean():.2f}")
print(f"Post-Crisis Sample Size: {len(post_crisis)} months | Mean Price: INR {post_crisis.mean():.2f}")
print(f"Calculated T-Statistic: {t_stat:.4f}")
print(f"Calculated P-Value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print(f"\nConclusion: Reject the Null Hypothesis (p < {alpha}). The price drop is statistically significant.")
else:
    print(f"\nConclusion: Fail to Reject the Null Hypothesis (p >= {alpha}). No statistically significant shift detected.")
````

##### Which statistical test have you done to obtain P-Value?

I performed an Independent Two-Sample T-test (specifically Welch's T-test variation) to calculate the exact $p$-value. The test evaluated whether the distinct mean closing valuations calculated across our split pre-crisis and post-crisis sample frames diverged by a statistically significant margin.

##### Why did you choose the specific statistical test?

I selected Welch's T-test because we are comparing the means of two entirely separate historical periods (independent groups) where the sample sizes and underlying variance parameters are completely unequal.As our descriptive statistics and boxplot visualizations previously verified, the pre-2018 period possesses massive, wide price distributions, whereas the post-2018 period features a completely flattened variance space. Standard Student's $t$-tests assume equal variances and would yield a biased result here. Welch's adaptation mathematically corrects for variance inequalities, guaranteeing that our resulting $p$-value remains precise, reliable, and free from structural bias.

Hypothetical Statement - 2: The historical monthly closing prices of Yes Bank are heavily skewed and deviate significantly from a standard Gaussian (normal) distribution profile.

##### Which statistical test have you done to obtain P-Value?

While our early diagnostic charts (the original Histogram, the Q-Q plot, and the heavily separated boxplots) strongly suggested a right-skewed tail, visual interpretations alone are not enough to justify data transformation steps to an interview panel. ### ML Model - 2

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

# Step 1: Initialize and fit the baseline regularized model
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train)

# Step 2: Generate prediction matrices (on log10 scale)
y_train_pred_ridge_log = ridge_model.predict(X_train_scaled)
y_test_pred_ridge_log = ridge_model.predict(X_test_scaled)

# Step 3: Inverse-transform back to raw INR values (10**y) for realistic reporting
y_train_raw = 10 ** y_train
y_test_raw = 10 ** y_test
y_train_pred_ridge_raw = 10 ** y_train_pred_ridge_log
y_test_pred_ridge_raw = 10 ** y_test_pred_ridge_log

def compute_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return [mae, rmse, r2, mape]

# Extract out metrics
ridge_metrics_train = compute_metrics(y_train_raw, y_train_pred_ridge_raw)
ridge_metrics_test = compute_metrics(y_test_raw, y_test_pred_ridge_raw)

# Store inside a dataframe for easy charting
ridge_score_df = pd.DataFrame({
    'Metric': ['MAE (INR)', 'RMSE (INR)', 'R2 Score', 'MAPE (%)'],
    'Train Score': ridge_metrics_train,
    'Test Score': ridge_metrics_test
})
print(ridge_score_df)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

ML Model - 2 is a regularized Ridge Regression model. Unlike our initial Ordinary Least Squares baseline, Ridge forces a squared magnitude penalty ($L_2$ regularization, controlled by hyperparameter $\alpha$) directly onto the linear regression coefficients. This structural optimization shrinks ballooning coefficients towards zero, which heavily dampens the multicollinearity variance amplification typical of raw stock indicators (Log_Open, Log_High, Log_Low).Performance Summary:Ridge displays high performance consistency between training and test boundaries. Because the $L_2$ penalty forces the weights of highly correlated features to distribute evenly, the model experiences less out-of-sample scaling drift compared to our unregularized baseline. The resulting test errors (MAE and RMSE) remain tightly bounded, establishing Ridge as a far safer, less volatile linear predictor for real-world execution.
